# performing PCA

```python
from sklearn.preprocessing import StandardScaler
stand_x=StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca_data=pca.fit_transform(stand_x)
```

```python
pca_data.shape
```

```python
pca_data.T
```

```python
pca_data=np.vstack((pca_data.T,y)).T
```

```python
pca_df=pd.DataFrame(data=pca_data,columns=["first_principal","second_principal","label"])
```

```python
pca_df
```

```python
import seaborn as sns
sns.FacetGrid(pca_df,hue="label",size=6).map(plt.scatter,"first_principal","second_principal").add_legend()
```

```python
pca_2=PCA(n_components=784)
pca_data=pca_2.fit_transform(stand_x)
```

```python
per_var_explained=pca_2.explained_variance_/np.sum(pca_2.explained_variance_)
```

```python
cum_var_explained=np.cumsum(per_var_explained)
```

```python


plt.plot(cum_var_explained,linewidth=2)

plt.grid()
plt.xlabel("n_components")
plt.ylabel("cummulative_explained_variance")
plt.show()
```

# performing tsne 

```python
from sklearn.manifold import TSNE
```

```python
model=TSNE(n_components=2,random_state=0)
tsne_data=model.fit_transform(stand_x)
```

```python
tsne_data
```

```python
tsne_data = np.vstack((tsne_data.T,y)).T
```

```python
tsne_df=pd.DataFrame(data=tsne_data,columns=("Dim_1","Dim_2","label"))
```

```python
tsne_df
```

```python
sns.FacetGrid(tsne_df,hue="label",size=6).map(plt.scatter,"Dim_1","Dim_2").add_legend()
plt.show()
```

```python
model = TSNE(n_components=2, random_state=0, perplexity=50)
tsne_data = model.fit_transform(stand_x) 


tsne_data = np.vstack((tsne_data.T,y)).T
tsne_df = pd.DataFrame(data=tsne_data, columns=("Dim_1", "Dim_2", "label"))


sns.FacetGrid(tsne_df, hue="label", size=6).map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
plt.title('With perplexity = 50')
plt.show()
```

```python
model = TSNE(n_components=2, random_state=0, perplexity=50,  n_iter=5000)
tsne_data = model.fit_transform(stand_x) 


tsne_data = np.vstack((tsne_data.T,y)).T
tsne_df = pd.DataFrame(data=tsne_data, columns=("Dim_1", "Dim_2", "label"))


sns.FacetGrid(tsne_df, hue="label", size=6).map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
plt.title('With perplexity = 50, n_iter=5000')
plt.show()
```

```python
model = TSNE(n_components=2, random_state=0, perplexity=2)
tsne_data = model.fit_transform(stand_x) 


tsne_data = np.vstack((tsne_data.T,y)).T
tsne_df = pd.DataFrame(data=tsne_data, columns=("Dim_1", "Dim_2", "label"))


sns.FacetGrid(tsne_df, hue="label", size=6).map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
plt.title('With perplexity = 2')
plt.show()
```
